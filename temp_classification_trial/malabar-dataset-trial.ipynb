```python
from keras.models import Model
from keras.layers import (
    Input, Conv2D, SeparableConv2D, MaxPooling2D, 
    GlobalAveragePooling2D, BatchNormalization, Activation, 
    Dense, Dropout, Add, Rescaling
)
from keras.optimizers import Adam
from keras import mixed_precision

# ✅ Enable mixed precision for faster GPU training (optional)
mixed_precision.set_global_policy('mixed_float16')

def Xception(input_shape=(299, 299, 3), classes=3):
    img_input = Input(shape=input_shape)

    # Normalize input to [0,1]
    x = Rescaling(1./255)(img_input)

    # ===== Entry Flow =====
    x = Conv2D(32, (3,3), strides=(2,2), use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(64, (3,3), use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # First block
    residual = Conv2D(128, (1,1), strides=(2,2), padding='same', use_bias=False)(x)
    residual = BatchNormalization()(residual)

    x = SeparableConv2D(128, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = SeparableConv2D(128, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)

    x = MaxPooling2D((3,3), strides=(2,2), padding='same')(x)
    x = Add()([x, residual])

    # Second block
    residual = Conv2D(256, (1,1), strides=(2,2), padding='same', use_bias=False)(x)
    residual = BatchNormalization()(residual)

    x = Activation('relu')(x)
    x = SeparableConv2D(256, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = SeparableConv2D(256, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)

    x = MaxPooling2D((3,3), strides=(2,2), padding='same')(x)
    x = Add()([x, residual])

    # Third block
    residual = Conv2D(728, (1,1), strides=(2,2), padding='same', use_bias=False)(x)
    residual = BatchNormalization()(residual)

    x = Activation('relu')(x)
    x = SeparableConv2D(728, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = SeparableConv2D(728, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)

    x = MaxPooling2D((3,3), strides=(2,2), padding='same')(x)
    x = Add()([x, residual])

    # ===== Middle Flow (8 times) =====
    for _ in range(8):
        residual = x
        # 1x1 conv to match channels if needed
        if residual.shape[-1] != 728:
            residual = Conv2D(728, (1,1), padding='same', use_bias=False)(residual)
            residual = BatchNormalization()(residual)

        x = Activation('relu')(x)
        x = SeparableConv2D(728, (3,3), padding='same', use_bias=False)(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        x = SeparableConv2D(728, (3,3), padding='same', use_bias=False)(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        x = SeparableConv2D(728, (3,3), padding='same', use_bias=False)(x)
        x = BatchNormalization()(x)

        x = Add()([x, residual])

    # ===== Exit Flow =====
    residual = Conv2D(1024, (1,1), strides=(2,2), padding='same', use_bias=False)(x)
    residual = BatchNormalization()(residual)

    x = Activation('relu')(x)
    x = SeparableConv2D(728, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = SeparableConv2D(1024, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((3,3), strides=(2,2), padding='same')(x)
    x = Add()([x, residual])

    x = SeparableConv2D(1536, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = SeparableConv2D(2048, (3,3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # ===== Classifier =====
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.5)(x)
    x = Dense(classes, activation='softmax', dtype='float32')(x)

    model = Model(img_input, x, name='xception_custom')
    return model


# ===== Compile Model =====
model = Xception(input_shape=(299, 299, 3), classes=3)
model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# View model summary
model.summary()
```
